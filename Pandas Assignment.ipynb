```python
car_sales_missing_data = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing_data
```

```python
car_sales_missing_data['Odometer'].fillna(car_sales_missing_data['Odometer'].mean(), inplace=True)
car_sales_missing_data
```

```python
car_sales_missing_data['Seats'] = 5
car_sales_missing_data
```

```python
import numpy as np
np.random.seed(365)
data = np.round(np.random.uniform(1.3, 4.5,size=len(car_sales_missing_data)),2)
car_sales_missing_data['Engine Size'] = data
car_sales_missing_data
```

```python
car_sales_missing_data['Price'] = car_sales['Price'].replace('[\$\,\.]', '', regex=True).astype(int) / 100
car_sales_missing_data['Price per KM'] = car_sales_missing_data['Price'] / car_sales_missing_data['Odometer']
car_sales_missing_data
```

```python
car_sales_missing_data.drop(columns=['Price per KM'],inplace=True)
car_sales_missing_data
```

```python
shuffled_car_sales_missing_data = car_sales_missing_data.sample(frac = 1)
shuffled_car_sales_missing_data
```

```python
shuffled_car_sales_missing_data.reset_index(inplace=True)
shuffled_car_sales_missing_data
```

```python
shuffled_car_sales_missing_data["Odometer"] = shuffled_car_sales_missing_data["Odometer"].apply(lambda x: x/1.6)
shuffled_car_sales_missing_data
```

```python
shuffled_car_sales_missing_data.rename(columns={'Odometer':'Odometer Miles'},inplace=True)
shuffled_car_sales_missing_data
```
